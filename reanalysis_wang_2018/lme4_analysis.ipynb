## Model comparison

### RT

```r
# model comparison for different random effects structures
rt_aics <- AIC(rt_bothNT,rt_bothNT2,rt_bothNT3,rt_bothNT4,rt_bothNT5,rt_bothNT6,rt_bothNT7,rt_bothNT8, rt_bothNT9)
rt_aics <- rt_aics[with(rt_aics, order(AIC)), ]
print(rt_aics)
```
